```python
# imports
import numpy as np
from matplotlib import pyplot as plt

from astropy.table import Table
```

## Check the output

```python
nlam = 240
nx = 100
ny = 100
lam0 = 2.77e3
dlam = 2.5e-1
dx = 0.4 # kpc
```

```python
spec_tbl = Table.read('../../HydroSims/FirstRun/spec_const.ascii', format='ascii')
```

```python
spec_tbl[0:5]
```

### Build spectrum

```python
counts = spec_tbl['col1'].data
```

```python
flux = np.zeros(nlam)
```

```python
for kk in range(nlam):
    tmp = np.roll(counts,-1*kk)[::100*100]
    flux[kk] = np.sum(tmp)
```

```python
wave = lam0 + dlam*np.arange(nlam)
```

### Plot

```python
plt.clf()
ax = plt.gca()
ax.plot(wave, flux)
plt.show()
```
